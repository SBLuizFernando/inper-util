## Acionamentos elétricos

![title](acionamentos-eletricos-criterios-avaliacao.png)

1. Serão realizadas quatro provas ao longo do semestre, A1, A2 (na semana de avaliações intermediárias), A3 e A4 (na semana de avaliações finais). A nota média destas quatro provas (MPRV) equivalerá a 45% da média final.
2. Serão divulgadas quatro listas de exercícios ao longo do semestre, EX1, EX2, EX3 e EX4, devendo ser submetidas via blackboard dentro do prazo estipulado. A nota média destas quatro listas de exercícios (MEX) equivalerá a 5% da média final.
3. Serão realizados dois projetos, PRJ1 e PRJ2, compostos por uma apresentação e arguição em sala de aula (Studio/Apresentação) e relatório a ser submetido pelo Blackboard. A nota média de projetos (MPRJ) equivalerá a 50% da média final.
4. Há ainda uma avaliação de participação nas aulas, medida por meio de quizzes no início das aulas e conclusão de outros exercícios. Essa nota não tem impacto direto na média final, mas pode abrandar os critérios de aprovação (ver orientação 5b)
5. Para ser considerado aprovado na disciplina o aluno deverá atender aos seguintes critérios:<br>
    5.a. Obter média de projetos, MPRJ igual ou superior a 5.0<br>
    5.b. Obter coeficiente 0,9*MPRV+0,1*MEX+0,04*PART igual ou superior a 5,0<br>
    5.c. O aluno que atender aos dois critérios (5a e 5b) terá sua média calculada conforme os pesos já apresentados.<br>
    5.d. O aluno que não satisfazer qualquer um dos critérios (5a) ou (5b) será considerado reprovado. Para fins de registro, sua média será a menor dentre os dois critérios (5a ou 5b).

```python
#Script de cáclulo de aprovação em Acionamentos elétricos

avaliacao_1 = 5
avaliacao_2 = 5
avaliacao_3 = 5
avaliacao_4 = 5
lista_1 = 5
lista_2 = 5 
lista_3 = 5
lista_4 = 5
projeto_1 = 5 
projeto_2 = 5
participacao = 5 


nota_ac_el = (avaliacao_1*11.25+avaliacao_2*11.25+avaliacao_3*11.25\
+avaliacao_4*11.25+lista_1*1.25+lista_2*1.25+lista_3*1.25+lista_4*1.25+\
projeto_1*25+projeto_2*25+participacao*0)/100

media_projetos = (projeto_1 + projeto_2)/2
media_provas = (avaliacao_1+avaliacao_2+avaliacao_3+avaliacao_4)/4
media_listas = (lista_1+lista_2+lista_3+lista_4)/4

coeficiente = 0.9*media_provas+0.1*media_listas+0.4*participacao

if media_projetos >=5 and coeficiente >= 5:
    print("Aprovado em Acionamentos Elétricos com média final de", nota_ac_el)
elif media_projetos < 5 and coeficiente >=5:
    print("Dependência de Acionamentos Elétricos pela média dos projetos de", media_projetos,\
          "menor que 5 e média final de", nota_ac_el)
elif coeficiente < 5 and media_projetos >= 5:
    print("Dependência de Acionamentos Elétricos pelo coeficiente de", coeficiente,\
          "menor que 5 e média final de", nota_ac_el)
elif media_projetos < 5 and coeficiente < 5:
    print("Dependência de Acionamentos Elétricos pela média de projetos de", media_projetos,\
          "menor que 5, pelo coeficiente de", coeficiente,"e média final de", nota_ac_el)
    
print("\nMédia final:", nota_ac_el)
print("Média de projetos:", media_projetos)
print("Coeficiente:", coeficiente)
```

## Ciência dos dados

![title](ciencia-dados-criterios-avaliacao.png)

AVALIAÇÃO:
A disciplina tem três avaliações e dois projetos e várias APSs, os quais possuem os seguintes critérios:

• Avaliação Inicial, Avaliação Intermediária e Avaliação Final:
As avaliações em Ciência dos dados serão aplicadas de acordo com as datas previstas pelo calendário do Insper. As notas das três avaliações serão todas expressas numericamente variando de 0 a 10 de acordo com o desempenho de cada aluno.

• Projeto 1 e Projeto 2:
Os projetos em Ciências dos dados estão previamente agendados no plano de aula. A presença desses projetos tem como propósito aprimorar os objetivos de aprendizagem destacados anteriormente em Competências Transversais. Dessa forma, o aluno deve demonstrar bom conhecimento e dedicação em todos os projetos, sejam esses individuais ou em grupo. Todos os projetos são importantes para o aluno adquirir conhecimento e atingir os objetivos de aprendizado dessa disciplina.
As notas individuais dos projetos serão expressas em conceito variando entre A+ a I de acordo com o desempenho de cada aluno ou grupo. Para fins de média, são considerados os seguintes valores de conversão: I=0; D=2,5; C=5; C+=6; B=7; B+=8; A=9; A+=10.

• Atividade prática supervisionada – APS:<br>
As Atividades Práticas Supervisionadas estão previamente programadas no plano de aula e devem ser feitas em horário extraclasse. Incluem, mas não se restringem a APS de Check de Projetos e APS de exercícios.

• Média Final:<br>
Para nota das avaliações (NA), será considerada a seguinte ponderação:<br>
        NA=20%Inicial+30%Intermediária+50%Final<br>
Para nota dos projetos (NP), será considerada a seguinte ponderação:<br>
        NP=10%Média das APSs+45%Projeto1+45%Projeto2<br>
A nota final da disciplina será calculada da seguinte forma:<br>
        média(NA,NP), se NA e NP forem maiores ou iguais a 5 simultaneamente;<br>
        min(NA,NP), caso contrário<br>

IMPORTANTE:
* Para os projetos, será utilizada a tabela oficial do Blackboard para converter de conceito para nota numérica.
* Na Média das APSs, serão consideradas todas as APSs disponilizadas.
* É preciso ter pelo menos 50% das APSs entregues para o critério da nota final acima ser aplicada; caso contrário, aluno será reprovado. 

```python
#Script de cáclulo de aprovação em Ciência dos Dados
avaliacao_inicial = 5
avaliacao_intermediaria = 5 #
avaliacao_final = 5

lista_aps = [5, 5, 5, 5, 5] #notas nas APSs 1, 2, 3, 4, ..., se não entregou alguma coloque 0
num_aps = 5 #número TOTAL de APSs que deveriam ser entregues
media_aps = sum(lista_aps)/num_aps

projeto_1 = 5 #C
projeto_2 = 5 #C

nota_avaliacoes = (20*avaliacao_inicial+30*avaliacao_intermediaria+50*avaliacao_final)/100
nota_projetos = (10*media_aps+45*projeto_1+45*projeto_2)/100

if nota_avaliacoes >= 5 and nota_projetos >= 5:
    nota_final = (nota_avaliacoes+nota_projetos)/2
else:
    nota_final = min(nota_avaliacoes, nota_projetos)
    
if len(lista_aps)>=(num_aps/2):
    passou_aps = True
else:
    passou_aps = False
    print("Dependência em Ciência dos Dados por não entregar pelo menos 50% das APSs.")
    
if nota_final >= 5 and passou_aps == True:
    print("Aprovado em Ciência dos Dados com média final de", nota_final)
elif nota_avaliacoes < 5 and nota_projetos >=5:
    print("Dependência em Ciência dos Dados por nota de avalições", nota_avaliacoes, "ser menor que 5.\
    com média de", nota_final)
elif nota_avaliacoes >= 5 and nota_projetos < 5:
    print("Dependência em Ciência dos Dados por nota de projetos", nota_avaliacoes, "ser menor que 5.\
    com média de", nota_final)
elif nota_avaliacoes < 5 and nota_projetos < 5:
    print("Dependência em Ciência dos Dados por nota de avalições", nota_avaliacoes,\
    "e média de projetos", nota_projetos, "serem menores que 5 e com média final de", nota_final)
    
print("\nMédia final:", nota_final)    
print("Nota de avaliações:", nota_avaliacoes)
print("Nota de projetos:", nota_projetos)
```

## Codesign de aplicativos

![title](codesign-aplicativos-criterios-avaliacao.png)

AVALIAÇÃO: Cada objetivo está associado a um ou mais instrumentos de avaliação.
Um objetivo pode ser avaliado por múltiplos instrumentos e um instrumento pode
avaliar múltiplos objetivos. Cada associação é medida por uma rubrica e resumida
em um conceito. Para fins de média e mediana, são considerados os seguintes
valores de conversão: I=0; D=2,5; C=5; C+=6; B=7; B+=8; A=9; A+=10. Para ser
aprovado na disciplina, um aluno deve satisfazer os seguintes critérios: (A)
para cada objetivo, a mediana dos conceitos associados a ele deve ser maior ou
igual a 4,5; (B) a Média Essencial, uma média ponderada de todos os conceitos,
independente do objetivo, deve ser maior ou igual a 5,0. A disciplina também
possui instrumentos de diagnóstico e a Média Complementar é uma média ponderada
das consolidações desses instrumentos. Por fim, a Média Final é uma combinação
da Média Essencial e da Média Complementar, mas condicionada aos critérios de
aprovação: se A não for satisfeito, a Média Final é o mínimo entre 4,0 e a Média
Essencial; se A for satisfeito mas B não for satisfeito, a Média Final é igual à
Média Essencial; se tanto A quanto B forem satisfeitos, a Média Final é o mínimo
entre 10,0 e uma soma de 100% da Média Essencial e 10% da Média Complementar.

### Matemática da variação

![title](matematica-varaiacao-criterios-avaliacao.png)

Critérios de aprovação<br>
A aprovação na disciplina será baseada nos elementos a, b e c descritos a seguir.<br>
a) Média das avaliações individuais (MI)<br>
Em cada avaliação individual, o aluno receberá uma nota numérica de 0 a 10.
Fazendo a média das notas numéricas das avaliações individuais AI1 até AI5, considerando os pesos indicados na tabela 

“Critérios de avaliação”, chega-se à média parcial MP, relativa aos 5 instrumentos de avaliação individuais.
• Se a MP for maior ou igual a 6,5, o aluno fica dispensado do Exame Final, ficando MI = MP.
• Se a MP for menor do que 6,5, o aluno deverá fazer o Exame Final (EF). Nesse caso, MI = 0,6*MP + 0,4*EF.

b) Avaliação do OA5 – APS de estudo autônomo de temas matemáticos <br>
A nota da APS, a qual reflete o estudo autônomo de tema não abordado em sala de aula, será obtida em avaliação específica para este fim.

c) Avaliação dos demais objetivos de aprendizagem <br>
A partir dos instrumentos de avaliação da disciplina, o aluno receberá um conceito em cada objetivo de aprendizagem, que reflete a rubrica apresentada ao final deste plano de aula. Tal conceito é alterado sempre que o aluno melhorar seu desempenho nas próximas avaliações ou nas Verificações Delta.

Aprovação na disciplina<br>
Para ser aprovado, além da frequência superior a 75%, o aluno precisa, necessariamente, satisfazer as duas condições abaixo. 
1. Ter média MI maior ou igual a 5,0.
2. Atingir pelo menos conceito C em todos os objetivos de aprendizagem.

Cálculo da média final<br>
a) Para alunos aprovados na disciplina e alunos que não cumprirem a condição 1 nem a 2, a média final M será calculada de acordo com a seguinte fórmula:<br>
M=0,9*MI+0,1*APS<br>
b) Para alunos que não cumprirem apenas a condição 1, a média final será igual a MI.<br>
c) Para alunos que não cumprirem apenas a condição 2, a média final será equivalente ao conceito do objetivo com pior desempenho.

```python
#Script de cáclulo de aprovação em Matemática da Variação
avaliacao_individual_1 = 5
avaliacao_individual_2 = 5
avaliacao_individual_3 = 5
avaliacao_individual_4 = 5
avaliacao_individual_5 = 5

media_parcial = (avaliacao_individual_1+avaliacao_individual_2+avaliacao_individual_3+avaliacao_individual_4+\
                avaliacao_individual_5)/5

aps_estudo_individual_OA5 = 5 #Escolher um tema e fazer uma prova sobre ele

objetivos_aprendizagem = ["C", "C", "C", "C", "C", "C"] #Objetivos de aprendizagem possíveis de recuperar com as deltas
        
if media_parcial >= 6.5:
    print("Você está dispensado do Exame Final")
    media_avaliacoes_individuais = media_parcial
elif media_parcial < 6.5:
    print("Você terá que fazer o Exame Final")
    nota_exame_final = float(input("Digite a nota do seu Exame Final: "))
    media_avaliacoes_individuais = 0.6*media_parcial+0.4*nota_exame_final
    
if media_avaliacoes_individuais>=5 and objetivos_aprendizagem.count("D") == 0:
    media_final = 0.9*media_avaliacoes_individuais+0.1*aps_estudo_individual_OA5
    if media_final>=5:
        print("\nAprovado em Matemática da Variação com media final de: ", media_final)
    else:
        print("\nDependência em Matemática da Variação por media final de: ", media_final, "ser menor que 5.")
        
elif media_avaliacoes_individuais<5 and objetivos_aprendizagem.count("D")>0:
    media_final = 0.9*media_avaliacoes_individuais+0.1*aps_estudo_individual_OA5
    if media_final>=5:
        print("\nAprovado em Matemática da Variação com media final de: ", media_final)
    else:
        print("\nDependência em Matemática da Variação por media final de: ", media_final, "ser menor que 5.")
        
elif media_avaliacoes_individuais<5 and objetivos_aprendizagem.count("D") == 0:
    media_final = media_avaliacoes_individuais
    if media_final>=5:
        print("\nAprovado em Matemática da Variação com media final de: ", media_final)
    else:
        print("\nDependência em Matemática da Variação por media final ser igual a média de avaliações individuais:",\
              media_final, "ser menor que 5.")
          
elif media_avaliacoes_individuais>=5 and objetivos_aprendizagem.count("D")>0:
    media_final = 2.5 #conversão do D de acordo com o Black Board
    if media_final>=5:
        print("\nAprovado em Matemática da Variação com media final de: ", media_final)
    else:
        print("\nDependência em Matemática da Variação por media final ser igual ao conceito do obejtivo\
        de aprendizagem com o pior desempenho: ", media_final, " ser menor que 5.")
         
            
print("\nMédia final:", media_final)
print("Conceitos dos objetivos de aprendizagem:", objetivos_aprendizagem)
print("Nota de APS - Estudo individual:", aps_estudo_individual_OA5)
```
